# Catboost

```python
from catboost import CatBoostRegressor 
from sklearn.model_selection import StratifiedKFold, GridSearchCV
import numpy as np
# Parameter grid
# params = {
#     'learning_rate': [0.1,0.08,0.05,0.09],
#     'depth': [8, 6, 10,12],
#     'l2_leaf_reg': [1, 3, 5]
# }

# Model
catb = CatBoostRegressor(learning_rate=0.09,depth=12,silent=True) 

# Evaluate
skf = StratifiedKFold(n_splits=10,shuffle=True, random_state=42)
scores = []

for k, (train, test) in enumerate(skf.split(X_train, y_train)):
    catb.fit(X_train.iloc[train], y_train.iloc[train])  
    y_pred = catb.predict(X_train.iloc[test])
    rmse = np.sqrt(mean_squared_error(y_train.iloc[test], y_pred))
    scores.append(rmse)
    print(f"Fold {k+1} RMSE: {rmse}")

print(f"\nMean RMSE: {np.mean(scores)}")

# Final evaluation
y_pred = catb.predict(X_test)
cat_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Final RMSE:", cat_rmse)
```
